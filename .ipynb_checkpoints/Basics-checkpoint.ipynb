```python
import numpy as np
import csv
with open("data/movies.txt", encoding="latin_1") as f:
    reader = csv.reader(f, delimiter="\t")
    
    # Put the contents in a matrix
    movies = np.array(list(reader))
data = np.loadtxt("data/data.txt")
```

```python
import seaborn as sns
ax = sns.distplot(data[:, 2], kde = False)
ax.set_ylabel("Occurences of Rating")
ax.set_xlabel("Rating")
ax.set_title("Occurences of Each Rating in Dataset")
ax.figure
```

```python
from collections import Counter
import pandas as pd
# Collect most popular movies
most_popular = Counter(data[:, 1]).most_common(10)
movie = []
num_ratings = []

for x, y in most_popular:
    movie.append(movies[int(x) - 1, 1])
    num_ratings.append(y)
plot_data = pd.DataFrame({"Movie" : movie, "Number of Ratings" : num_ratings})
```

```python
import matplotlib.pyplot as plt
plt.bar(x = movie, height = num_ratings, width = 1)
plt.xticks(rotation=90)
plt.xlabel("Movie")
plt.ylabel("Number of Ratings")
plt.title("Top 10 Most Popular Movies")
plt.show()
```

```python
count_ratings = Counter(data[:, 1])
rating_sum = {}
for u, m, rating in data:
    if rating_sum.get(m, -1) == -1:
        rating_sum[m] = rating
    else:
        rating_sum[m] += rating
for key in rating_sum:
    rating_sum[key] = rating_sum[key] / count_ratings[key]
```

```python
import operator
# Get top 10 movies with highest average rating
sorted_ratings = sorted(rating_sum.items(), key=operator.itemgetter(1))[-10:]
```
